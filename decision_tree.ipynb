$P(FraudFoundP) = sum(FraudFoundP) / count(FraudFoundP)$

```python
# get the percentage of cases where the target variable was found to be 1 (fraud was found)
probFraudFound = sum(data.loc[:,"FraudFound_P"]) / data.loc[:,"FraudFound_P"].count()
probFraudNotFound = (data.loc[:,"FraudFound_P"].count() - sum(data.loc[:,"FraudFound_P"])) / data.loc[:,"FraudFound_P"].count()
print(probFraudFound)
```

$Entropy = E(X) = - \sum_{i=1}^n P(x_i)log_2 P(x_i)$

$E(FraudFoundP) = E(Dataset) = -P(FraudFoundP) * log_2 P(FraudFoundP) - P(!FraudFoundP) * log_2 P(!FraudFoundP)$


```python
# calculate the entropy of fraud found, which is equivalent to the entropy of the dataset
entropyFraudFound = -probFraudFound * np.log2(probFraudFound) - probFraudNotFound * np.log2(probFraudNotFound)
print(entropyFraudFound)
```
